Using a conditional, carete a function called **relu** that takes an input x and returns a values. The function will be defined as follows:

relu(x)=x,

if x≥0, relu(x)=x, 0, otherwise

Test your function as follows to see whether it returns the expected value or not.  e.g **relu(-3)**


```python
def relu(x):
    if x >= 0:        # if x is zero or positive
        return x      # return x as it is (pass it through unchanged)
    else:             # if x is negative
        return 0      # squash it to 0

print(relu(5))   # testing with a positive number - should return 5
print(relu(-3))  # testing with a negative number - should return 0

```

**Exercise 1.2**

Write a Python function to find the **Max of two numbers** and another one to find the **Max of three numbers**.

```python
def max_of_two(a, b):
    if a > b:       # if a is bigger than b
        return a    # return a as it is the largest
    else:           # otherwise b is bigger or they are equal
        return b    # return b as it is the largest

print(max_of_two(3, 7))   # should print 7
print(max_of_two(10, 4))  # should print 10
print(max_of_two(5, 5))   # should print 5 - equal numbers

def max_of_three(a, b, c):
    if a > b and a > c:   # if a is bigger than both b and c
        return a          # a is the largest
    elif b > a and b > c: # if b is bigger than both a and c
        return b          # b is the largest
    else:                 # otherwise c must be the largest
        return c          # c is the largest

print(max_of_three(3, 7, 10))   # should print 10
print(max_of_three(10, 4, 8))   # should print 10
print(max_of_three(5, 5, 5))    # should print 5
```

Numpy is the core library for scientific computing in Python. It provides a high-performance multidimensional array object, and tools for working with these arrays.

**Create Array**

To use Numpy, we first need to import the numpy package:

```python
# define an array
import numpy
mylist = [[1, 2, 3], [3, 4, 5]]
myarray = numpy.array(mylist)
print(myarray)
print(myarray.shape)
```

Array notation and ranges can be used to efficiently access data in a NumPy array.

```python
# access values
print("First row: %s" % myarray[0])
print("Last row: %s" % myarray[-1])
print("Specific row and col: %s" % myarray[0, 2])
print("Whole col: %s" % myarray[:, 2])
```

**Exercise 2.5**

Write a Python script to calculate the following statistical values from a NumPy array of size 10 with random values between 1 and 50:

*   Standard deviation
*   Variance
*   Median
*   75th percentile

```python
import numpy as np

# create an array of 10 random integers between 1 and 50
a = np.random.randint(1, 51, size=10)
print("Array:", a)

# standard deviation - how spread out the numbers are from the average
print("Standard deviation:", np.std(a))

# variance - standard deviation squared, another measure of spread
print("Variance:", np.var(a))

# median - the middle value when sorted
print("Median:", np.median(a))

# 75th percentile - 75% of values fall below this number
print("75th percentile:", np.percentile(a, 75))
```

Create a 6x6 NumPy array with a checkerboard pattern (alternating 1s and 0s).

```python
import numpy as np

# create a 6x6 grid of zeros
board = np.zeros((6, 6), dtype=int)

# fill in the 1s at alternating positions to create checkerboard pattern
board[0::2, 0::2] = 1   # even rows, even columns get a 1
board[1::2, 1::2] = 1   # odd rows, odd columns get a 1

print(board)
```

**Exercise 2.6**

Create a 1D NumPy array ([10, -5, 3, -2, 15, -7]). Replace all negative values with 0, without modifying the original array.

```python
import numpy as np

# create the original array
original = np.array([10, -5, 3, -2, 15, -7])

# make a copy so the original is not modified
modified = original.copy()

# replace all negative values in the copy with 0
modified[modified < 0] = 0

# print both to prove the original is unchanged
print("Original:", original)   # should still have negatives
print("Modified:", modified)   # should have 0s instead of negatives
```
